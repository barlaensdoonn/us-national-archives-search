```python
import requests
from collections import Counter
```

```python
api_url = "https://www.loc.gov/collections/?fo=json"
r = requests.get(api_url)
collections = r.json()
```

```python
next_page = True
subjects = Counter()
titles = []
batch = 0

r = requests.get("https://www.loc.gov/collections/?fo=json")
collects = r.json()


while next_page:
    batch += 1
    print('getting results for batch {}'.format(batch))
    for collect in collects["results"]: 
        titles.append(collect["title"])
        try:
            for subject in collect['subject']:
                subjects[subject] += 1
        except KeyError:
            print('no subject field in this collection')
            subjects['none'] += 1
    next_page = collects["pagination"]["next"] #get the next page url
    if next_page: #make sure we haven't hit the end of the pages
        r = requests.get(next_page)
        collects = r.json()
    else:
          print("next_page is empty, looks like we're done")
```
